```python

plt.figure(figsize=(10, 5))
ax = sns.countplot(data=df, x='age_group', order=sorted(df['age_group'].unique()))
plt.title("Count of Patients per Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")

#Let's show the value of each group on our char
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2, height + 0.5,  
            int(height), ha='center')

plt.show()
```

```python
#Most patients are between the age of 21 to 40 years of age 
```

```python
#How many males and females are there
```

```python
plt.figure(figsize=(6, 4))
ax = sns.countplot(data=df, x='sex')
plt.title("Gender Distribution (0 = Female, 1 = Male)")

# Show the value of each group with 0 decimal points
for p in ax.patches:
    count = p.get_height()
    ax.annotate(f'{count:.0f}', (p.get_x() + p.get_width() / 2., count),
                ha='center', va='bottom', fontsize=12, color='black', fontweight='bold')

plt.show()
```

```python
#How common is each symptom
```

```python
symptom_cols = df.columns[3:-1] 
symptom_counts = df[symptom_cols].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
ax = sns.barplot(x=symptom_counts.index, y=symptom_counts.values, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Symptom Frequency Across All Patients")
plt.ylabel("Number of Patients")

#Let's show the value of each group on our chart
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2, height + 1, 
            int(height), ha='center')

plt.show()
```

```python
# Bivariate analysis Symptoms vs Severe Malaria
```

```python
severe_malaria = df[df['severity'] == 1]
non_severe = df[df['severity'] == 0]

symptom_severity = {}

for col in symptom_cols:
    severe_rate = severe_malaria[col].mean()
    non_severe_rate = non_severe[col].mean()
    symptom_severity[col] = [severe_rate, non_severe_rate]

symptom_df = pd.DataFrame(symptom_severity, index=['Severe', 'Non-Severe']).T.sort_values(by='Severe', ascending=False)

ax = symptom_df.plot(kind='bar', figsize=(14, 6))
plt.title("Symptom Rates in Severe vs Non-Severe Malaria Cases")
plt.ylabel("Proportion of Patients")
plt.xticks(rotation=45)
plt.tight_layout()

#Let's show the value of each group on our chart
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2, height + 0.01, f'{height:.2f}', ha='center')

plt.show()
```

```python
plt.figure(figsize=(14, 10))
sns.heatmap(df[symptom_cols.tolist() + ['severity']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Symptoms and Severity")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x='severity', y='age')
plt.title("Age Distribution by Malaria Severity (0 = Mild, 1 = Severe)")
plt.xlabel("Severe Malaria")
plt.ylabel("Age")
plt.show()
```

```python
group_severity = df.groupby('age_group')['severe_maleria'].mean().sort_index()

plt.figure(figsize=(10,5))
ax = group_severity.plot(kind='bar', color='teal')
plt.title("Average Severe Malaria Rate by Age Group")
plt.ylabel("Severity Rate")
plt.xlabel("Age Group")

# Lrt's add our values on each bar 
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2, height + 0.01, f'{height:.2f}', ha='center')

plt.show()
```

```python
severity_by_sex = df.groupby('sex')['severity'].mean()

ax = severity_by_sex.plot(kind='bar', color='orange')
plt.title("Severe Malaria Rate by Gender (0 = Female, 1 = Male)")
plt.ylabel("Severity Rate")

# values on each bar
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2, height + 0.01, f'{height:.2f}', ha='center')

plt.show()
```

```python
import math

#number of 0s and 1s
severity_counts = df['severity'].value_counts().sort_index()  # 0 = Negative, 1 = Positive

# bar and pie charts
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Bar chart with red and light green
severity_counts.plot(kind='bar', ax=axes[0], color=['red', 'lightgreen'])
axes[0].set_title('Severe Malaria Distribution')
axes[0].set_xlabel('Status')
axes[0].set_ylabel('Number of Patients')
axes[0].set_xticklabels(['Negative', 'Positive'], rotation=0)

# Add values on top of bars
for i, count in enumerate(severity_counts):
    axes[0].text(i, count + 1, str(count), ha='center')

#  round up percentage labels
def ceil_autopct(pct):
    return f'{math.ceil(pct)}%'

# Pie chart with red and light green
axes[1].pie(severity_counts, labels=['Negative', 'Positive'],
            autopct=ceil_autopct, colors=['red', 'lightgreen'])
axes[1].set_title('Severe Malaria (%)')

plt.tight_layout()
plt.show()
```
